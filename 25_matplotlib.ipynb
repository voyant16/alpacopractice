```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

```python
fig = plt.figure() #그림을 그리기 위한 도화지를 한장 준비해서 fig 변수에 할당

ax1= fig.add_subplot(1,1,1) #1행 1열에 1번째 자리
ax1.hist(tips['total_bill'], bins = 10) 
```

```python
#기존 그림 위에 막대기 20개로 덧그려줌(기본값보다 상세하게)
ax1.hist(tips['total_bill'], bins = 20) 

fig
```

```python
#기존 그림 위에 막대기 30개로 덧그려줌(기본값보다 상세하게)
ax1.hist(tips['total_bill'], bins = 30) 

fig
```

```python
fig = plt.figure() #그림을 그리기 위한 도화지를 한장 준비해서 fig 변수에 할당
ax1= fig.add_subplot(1,1,1)

ax1.scatter(tips['total_bill'], tips['tip'], c=tips['size'], s=tips['size']*20, alpha=0.5) 
#사이즈마다 다른 색으로, 동그라미 크기 다르게 표시 가능. 원래 size가 2 , 3, 4 이러니까 동그라미의 크기 차이를 키우려고 20씩 곱해준 것.
#alpha로 투명도 설정 가능
ax1.set_xlabel("TOTAL BILL") #x,y축 이름 정해줄 수 있음. 이렇게 matplot lib는 내가 원하는 대로 하나하나 다 설정 가능
ax1.set_ylabel("TIP") 
ax1.set_title("SCATTER")
```

```python
fig = plt.figure() #그림을 그리기 위한 도화지를 한장 준비해서 fig 변수에 할당
ax1= fig.add_subplot(1,1,1)

ax1.plot(tips['total_bill']) #선으로 그림
```

```python
fig = plt.figure() #그림을 그리기 위한 도화지를 한장 준비해서 fig 변수에 할당
ax1= fig.add_subplot(1,1,1)

male = tips[tips['sex'] == 'Male'] #팁스데이터의 성별이 남자인 것만 뽑고 싶다
female = tips[tips['sex'] == 'Female'] #여자인 것만

ax1.boxplot([male['total_bill'], female['total_bill']]) #상자그래프
```

```python
#그림을 하나만 그리고 싶다(도화지 할당 않고도. 다만 x축 y축 이름 할당 이런 거 어려움)

plt.scatter(tips['total_bill'], tips['tip'])
```
