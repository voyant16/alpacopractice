```python
import matplotlib.pyplot as plt
import seaborn as sns #데분 연습용 라브(연습용데이터, 가시화)
import pandas as pd
```

```python
#앤스컴: 평균, 표준편차, 상관계수, 회귀계수 다 똑같은 연습용 4개의 집단 제공
anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
anscombe[anscombe['dataset'] == 'I'] #앤스컴의 dataset 열이 1인 것에만 접근하고 싶어!
```

```python
data1 = anscombe[anscombe['dataset'] == 'I']
data2 = anscombe[anscombe['dataset'] == 'II']
data3 = anscombe[anscombe['dataset'] == 'III']
data4 = anscombe[anscombe['dataset'] == 'IV']

#data1~4를 다 나눠줌
```

```python
data3
```

```python
#네 개 데이터 평균값 다 보기 (다 똑같음)
print(data1.mean())
print(data2.mean())
print(data3.mean())
print(data4.mean())
```

```python
print(data1.std())
print(data2.std())
print(data3.std())
print(data4.std())
```

```python
fig = plt.figure() #그림을 그리기 위한 메인 도화지

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
#scatter: 산점도 그리기

ax1.scatter(data1['x'], data1['y'])
ax2.scatter(data2['x'], data2['y'])
ax3.scatter(data3['x'], data3['y'])
ax4.scatter(data4['x'], data4['y'])

fig #도화지를 보여줘!
```

```python
#plot: 선그래프. 뒤에 컬러도 설정 가능

ax1.plot(data1['x'], data1['y'], 'red')
ax2.plot(data2['x'], data2['y'], 'red')
ax3.plot(data3['x'], data3['y'], 'red')
ax4.plot(data4['x'], data4['y'], 'red')

fig
```

```python
ax1.set_title("Data1")
ax2.set_title("Data2")
ax3.set_title("Data3")
ax4.set_title("Data4")

fig
```

```python
fig.suptitle("Main Data")
fig
```
